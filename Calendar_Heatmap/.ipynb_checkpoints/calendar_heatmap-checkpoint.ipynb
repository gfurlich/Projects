```python
import calendar_heatmap as cal
from datetime import datetime
```

## Generate Example Data

```python
start_date = datetime.strptime('2017-10-01', '%Y-%m-%d')
end_date = datetime.strptime('2019-06-30', '%Y-%m-%d')

df = cal.generate_example_data(start_date, end_date, 0, 30)

df.head(10)
```

## Example Horizontal Calendar Heatmap

```python
f, ax = cal.calendar_heatmap(df, title='Horizontal Calendar Heatmap', cbar_title='h cbar', label='0211')
f.savefig('htest.png', bbox_inches='tight')
```

## Example Vertical Calendar Heatmap

```python
f, ax = cal.calendar_heatmap(df, title='Vertical Calendar Heatmap', cmap='Reds', cbar_title= 'v cbar', orientation='v')
f.savefig('vtest.png', bbox_inches='tight')
```
